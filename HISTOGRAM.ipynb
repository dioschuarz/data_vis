```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns

# Create a figure
fig, ax1 = plt.subplots(figsize=(16, 8), tight_layout=True)
fig.set_facecolor('white')

# Create a secondary axis
ax2 = ax1.twinx()

# Density Plot and Histogram of all arrival delays
sns.distplot(x = np.random.randn(100), hist=True, kde=False, 
             bins=int(50), color = 'darkred', 
             hist_kws={'edgecolor':'black'},
             ax=ax1)
sns.distplot(x = np.random.randn(100), hist=False, kde=True, 
             color = 'darkred',
             kde_kws={'linewidth': 4},
             ax=ax2)

# Create a line in graph
#plt.axvline([10], color='orange', linestyle = '--')
#plt.axvline([75], color='orange', linestyle = '--')

# Set title and labels
ax1.set_title('Histograma', size=30)
ax1.set_xlabel('Dados', size=15)
ax1.set_ylabel('Frequncia', size=15)
ax2.set_ylabel('Densidade', size=15)

# Set axis limits
#ax1.set_ylim([-10, 1])
#ax1.set_xlim([-10, 1])
#ax2.set_ylim([-10, 1])
#ax2.set_xlim([-10, 1])
```
